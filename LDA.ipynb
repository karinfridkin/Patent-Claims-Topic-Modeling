# Building LDA model

```python
#imports
import pandas as pd
import pickle
import gensim
import pyLDAvis.gensim
import warnings
```

```python
# Load data
data = pd.read_pickle('data_words.pkl')
corpus = pd.read_pickle('corpus.pkl')
```

```python
# Create Gensim Dictionary and Corpus
id2word = gensim.corpora.Dictionary(data)
num_corpus = [id2word.doc2bow(doc) for doc in data]
```

```python
# Save processed data
with open('num_corpus.pkl', 'wb') as f:
    pickle.dump(num_corpus, f)

with open('id2word.pkl', 'wb') as f:
    pickle.dump(id2word, f)
```

```python
# Create TF-IDF model
tfidf = gensim.models.TfidfModel(dictionary=id2word)
corpus_tfidf = tfidf[num_corpus]
```

```python
# Train LDA model
lda_model = gensim.models.LdaModel(corpus=corpus_tfidf,
                                    id2word=id2word,
                                    num_topics=10,
                                    random_state=100,
                                    update_every=1,
                                    chunksize=1,
                                    passes=10,
                                    alpha="auto")
```

```python
# Prepare LDAvis visualization for LDA model
pyLDAvis.enable_notebook()

vis_lda = pyLDAvis.gensim.prepare(lda_model, corpus_tfidf, id2word, mds="mmds", R=20)
```

```python
# Train another LDA model with different parameters
lda_model2 = gensim.models.LdaModel(corpus=corpus_tfidf,
                                    id2word=id2word,
                                    num_topics=10,
                                    random_state=42,
                                    update_every=1,
                                    chunksize=100,
                                    passes=10,
                                    alpha="auto")
```
